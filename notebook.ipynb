#### Import packages

```python
import pandas as pd
import dvc
import os
import io
```

#### Import data through its specific version

```python
# Set path to the working directory of your project
os.chdir(r"R:\projects\dsti_mlops")

# Specify the path to your wine_original.csv based on the working directory of your project
file_path = 'data_versioning/data/wine_original.csv'

# Specify a version of the wine_original you want to read, if not the latest one will be read
file_version = "v2" # None for latest version

try:
    with dvc.api.open(
        path=file_path,
        repo='.',  # Use '.' to indicate the current directory
        rev=file_version,
        mode='r'
    ) as f:
        data = pd.read_csv(io.StringIO(f.read()), sep=";")
    print(f"Successfully read data for version {file_version}")
except Exception as e:
    print(f"Error reading data for version {file_version}: {str(e)}")

data.head(n=10)
```
